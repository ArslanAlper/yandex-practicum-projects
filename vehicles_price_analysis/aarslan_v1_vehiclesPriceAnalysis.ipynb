```python
vehicles['model_year'].hist(bins=60, range=(1960, 2019), figsize=(15, 7))
plt.xlabel('\nModel Year', size=15, color='blue')
plt.ylabel('Vehicles\n', size=15, color='blue')
plt.xticks(color='b', size=12)
plt.yticks(color='b', size=12)
plt.title('\nDistribution of Vehicles by Model Years\n', size=21, color='blue');
```

```python
vehicles.plot(x='model_year', y='price', kind='scatter', alpha=.2, figsize=(15, 7))
plt.xlabel('\nModel Year', size=15, color='blue')
plt.ylabel('Price\n', size=15, color='blue')
plt.xticks(color='b', size=12)
plt.yticks(color='b', size=12)
plt.title('\nPrices by Model Years\n', size=21, color='blue');
```

General estimate:

1. We can conclude from that while the 'model_year' increases the price also increases. 

2. The second point is very important. Although there are less cars for 60's and 70's, those have same prices as much as new models, and we need to examine those from another aspect. 

3. The third point, there are some outliners both price and model_year columns. We will deal with those in the next steps.

```python
def find_mileage(row):
    year = row['age']
    km = row['odometer']
    month = row['month']
    age = year + (month/12)
    age = round(age, 2)
    yearly_mileage = km / age
    return int(round(yearly_mileage))
```

### Outlier detections

- We will examine the methods to detect outliers.

#### 1. 'IQR Method'

- We know **IQR, Interquartile Range**. The whisker plot computes the outlier ranges to the IQR. Let's define a function for that.

```python
def IQR_limits(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = 1.5 * IQR + Q3
    lower_limit = Q1 - 1.5 * IQR
    return [lower_limit, upper_limit]
```

```python
IQR_lim_prices = IQR_limits(vehicles.price)
display(IQR_lim_prices)
```

- According to the IQR, **the upper limit should be 34.750**. Let's examine another method.

#### 2. '3 Sigma Method'

We can use **3-sigma** as a reference to detect outliers. Let's write a function that returns the limit values.

```python
def sigma3_limits(column):
    mean = column.mean()
    sigma = column.std()
    upper_limit = mean + 3*sigma
    lower_limit = mean - 3*sigma
    return [lower_limit, upper_limit]
```

```python
sigma3_lim_prices = sigma3_limits(vehicles.price)
display(sigma3_lim_prices)
```

- According to the *3-sigma method*, **the upper limit is 42.321**. Let's look the 'z'-score graph

##### 'z' Score Graph on prices
