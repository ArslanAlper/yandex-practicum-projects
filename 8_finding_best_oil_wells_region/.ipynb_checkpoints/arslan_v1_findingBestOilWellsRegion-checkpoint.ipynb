# Project description

The task at hand is **to find the optimal location** for a new well for a mining company. To accomplish this, we will take the following steps:

- Collect data on oil well parameters such as oil quality and volume of reserves for the selected region.
- Build a predictive model to estimate the volume of reserves in the new wells.
- Select the oil wells with the highest estimated reserves.
- Evaluate the potential profit and risks of the selected oil wells using the Bootstrapping technique.
- Pick the region with the highest total profit.

We have access to data on oil samples **from three different regions**, including information on each oil well in those regions. The goal is to use this data to determine which region offers the highest profit margin.

# Project instructions

We will;

1- Download and prepare the data, explain the procedure.

2- Train and test the model for each region:

    2.1. Split the data into a training set and validation set at a ratio of 75:25.
    2.2. Train the model and make predictions for the validation set.
    2.3. Save the predictions and correct answers for the validation set.
    2.4. Print the average volume of predicted reserves and model RMSE.
    2.5. Analyze the results.

3- Prepare for profit calculation:
    
    3.1. Store all key values for calculations in separate variables.
    3.2. Calculate the volume of reserves sufficient for developing a new well without losses. Compare the obtained value with the average volume of reserves in each region.
    3.3. Provide the findings about the preparation for profit calculation step.

4- Write a function to calculate profit from a set of selected oil wells and model predictions:

    4.1. Pick the wells with the highest values of predictions.
    4.2. Summarize the target volume of reserves in accordance with these predictions
    4.3. Provide findings: suggest a region for oil wells' development and justify the choice. Calculate the profit for the obtained volume of reserves.

5- Calculate risks and profit for each region:

    5.1. Use the bootstrapping technique with 1000 samples to find the distribution of profit.
    5.2. Find average profit, 95% confidence interval and risk of losses. Loss is negative profit, calculate it as a probability and then express as a percentage.
    5.3. Provide findings: suggest a region for development of oil wells and justify the choice.

# Data description

Geological exploration data for the three regions are stored in files:

- 'geo_data_0.csv' .download dataset

- 'geo_data_1.csv' .download dataset

- 'geo_data_2.csv' .download dataset

- id — unique oil well identifier

- f0, f1, f2 — three features of points (their specific meaning is unimportant, but the features themselves are significant)

- product — volume of reserves in the oil well (thousand barrels)

# Conditions

- Only linear regression is suitable for model training (the rest are not sufficiently predictable).

- When exploring the region, a study of 500 points is carried with picking the best 200 points for the profit calculation.

- The budget for development of 200 oil wells is 100 USD million.

- One barrel of raw materials brings 4.5 USD of revenue. The revenue from one unit of product is 4,500 dollars (volume of reserves is in thousand barrels).

- After the risk evaluation, keep only the regions with the risk of losses lower than 2.5%. From the ones that fit the criteria, the region with the highest average profit should be selected.

The data is synthetic: contract details and well characteristics are not disclosed.

### Outlier Detection for all datasets

- First, we will define a function to find lower and upper limits to detect outliers.

```python
def IQR_limits(feature):
    Q1 = feature.quantile(0.25)
    Q3 = feature.quantile(0.75)
    IQR = Q3 - Q1
    upp_lim = 1.5 * IQR + Q3
    low_lim = Q1 - 1.5 * IQR
    return np.array([low_lim, upp_lim])
```

- We will find outlier limits and their amounts and percentages.

```python
for i, data in enumerate(datasets):
    
    print('\n\nOutlier Limit Values for geo_data_{}:\n'.format(i))
    for col in data.columns[1:]:
        print('| For {: <8} | lower limit : {: <6.2f} | upper limit : {: <6.2f} |'
              .format(col, IQR_limits(data[col])[0], IQR_limits(data[col])[1]))
    
    print('\nOutlier Counts for geo_data_{}:\n'.format(i))
    for col in data.columns[1:]:
        count_lowers = data.loc[data[col] < IQR_limits(data[col])[0]][col].count()
        count_uppers = data.loc[data[col] > IQR_limits(data[col])[1]][col].count()
        print('| For {: <8} | lower counts : {: <5} - {:.2%} | upper counts : {: <5} - {:.2%} |'
              .format(col, count_lowers, count_lowers/len(data), 
                      count_uppers, count_uppers/len(data)))
```

## Writing a function to calculate profit from a set of selected oil wells and model predictions

- We will define a function to get a profit Series from an oil well set.

```python
def profit(oil_wells):
    total_revenue = 0
    for reserve in oil_wells:
        revenue = reserve * product_price
        total_revenue += revenue
    profit = total_revenue - budget
    return profit
```

#### Hypotheses for Reserve Predictions

- We will compare the **predictions and target reserves for each region** with hypotheses tests.
- We will first run a **levene test** to compare the equality variances in the loop.
- Then we will run **ttest** based on the equality of variances.
- We will print the results.

#### Levene and T-Test for Predictions and Targets

- **H0:** 'The **average predictions** of related Region reserves **equals** to the **average target reserves** of related Region'
    (µ1 = µ2)
- **H1:** 'The **average predictions** of related Region reserves and the **average target reserves** of related Region are **different'**
    (µ1 < > µ2)
    
- alpha = 0.05

```python
alpha = 0.05

for i, y in enumerate(y_tests):
    
    results_levene = st.levene(preds_for_200[i], y)
    
    if (results_levene.pvalue < alpha):
        results = st.ttest_ind(preds_for_200[i], y, equal_var=False)
    
    else:
        results = st.ttest_ind(preds_for_200[i], y, equal_var=True)
    
    p_value = results.pvalue
    print("\np-value for Region {}: {}".format(i, p_value))
    
    if (p_value < alpha):
        print("We reject the null hypothesis")
        print("We assume that the avg. reserves of predictions and targets are different for Region {}\n".format(i))
    else:
        print("We can't reject the null hypothesis")
        print("We assume that the avg. reserves of predictions and targets are same for Region {}\n".format(i))
```

- We assume that the **average reserves of predictions and targets for each Region are different.**

#### Hypotheses Tests for Comparing of Regions

- **Region 1 has the highest reserve volumes as average.**
- We will make hypothesis test for Region 0 and Region 1.
- We consider the Region 1 have much more reserves than Region 0.

##### Levene Test

- **H0:** 'The variances of predictions for **Region 0** reserves and **Region 1** reserves are **equal.**'
    (var for Region 0 = var for Region 1)
- **H1:** 'The variances of predictions for **Region 0** reserves and **Region 1** reserves are **different.**'
    (var for Region 0 < > var for Region 1)
- **alpha = 0.05**

If the p-value is smaller than the alpha, then we assume the variances are different.

```python
alpha = 0.05

results = st.levene(preds0_for_200, preds1_for_200)

print('p-value', results.pvalue)

if (results.pvalue < alpha):
    print("Different variances assumed")
else:
    print("Equal variances assumed")
```

- We assume **variances are different**, so we will define that in t-test.

##### T-Test

- **H0:** 'The **average predictions of Region 1 reserves** equals to **the average predictions of Region 0 reserves'**
    (µ1 = µ2)
- **H1:** 'The average predictions of **Region 1 reserves is bigger** than the average predictions of Region 0 reserves'
    (µ1 > µ2)
    
- alpha = 0.05

- We will use the 'scipy.stat.ttest_ind' to test our hypotheses. The 'equal_var' parameter will be **'False'** since we assumed the variances are different. 
- If the p-value is bigger than the significance level (alpha), we will fail to reject the null hypothesis and assume the same expected values for the next year. Otherwise we will reject the null hypothesis.

```python
alpha = 0.05

results = st.ttest_ind(preds1_for_200, preds0_for_200, equal_var=False)
p_value = results.pvalue / 2

print("p-value: {}".format(p_value))
if (p_value < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

- We assumed that **the Region 1 has higher volume of predicted reserves than the Region 0 as average.**

### Using the bootstrapping technique with 1000 samples to find the distribution of profit

- We will use bootstrapping method to select **the best 200 oil wells from randomly selected 500 points**.
- We will select randomly 500 points as sample, and we will select the best 200 wells as subsample, then we will calculate profits **for target reserves.**
- We will make this step **1000 times** with for loop.
- We will save **profits of each Region in random_profits** to see the possible profit distribution.
- We will save **random_profits in random_profits_all**.

```python
random_profits_all = []  # We will use this list for storing the profit distributions of all regions.

for i, pred in enumerate(preds):
    
    random_profits = []  # We will store the 1000 total profits of each region into this list.
    state = np.random.RandomState(90)
    
    for k in range(1000):
        
        pred_sample = pred.sample(n=500, replace=True, random_state=state)  # Randomly selected 500 points predictions
        pred_subsample = pred_sample.sort_values(ascending=False)[:200]  # Best 200 oil wells from 500 points
        target_200 = y_tests[i][pred_subsample.index]  # Target values for subsample prediction
        total_profit = profit(target_200)  # Total profit of all 200 oil wells
        random_profits.append(total_profit)
    random_profits_all.append(pd.Series(random_profits))
```

```python
random_profits_0, random_profits_1, random_profits_2 = [p for p in random_profits_all]
random_profits_all = [random_profits_0, random_profits_1, random_profits_2]
```

### Finding average profit, 95% confidence interval and risk of losses. Loss is negative profit, calculate it as a probability and then expressing as a percentage

- We will find the Confidence Interval for each Region.

```python
for i, profs in enumerate(random_profits_all):
    ci = np.round(st.t.interval(0.95, (len(profs)-1), profs.mean(), profs.sem()), 2)
    print('95% confidence interval for Region {}: {}'.format(i, ci))
```

- Region 2 is the third.
- If we look at the distributions, we can see **the most risky** region as also **'Region 2'**.
- We will find the probabilities of loss (lower than 0) with **st.norm.cdf** function.

```python
loss_probs = []
threshold = 0
for profs in random_profits_all:
    mu = profs.mean()
    sigma = profs.std()
    prob = st.norm(mu, sigma).cdf(threshold)
    loss_probs.append(prob)
```
