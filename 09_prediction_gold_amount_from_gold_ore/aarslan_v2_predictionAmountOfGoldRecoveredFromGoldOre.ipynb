## Technological Process

Mined ore undergoes primary processing to get the ore mixture or rougher feed, which is the raw material for **flotation (also known as the rougher process)**. After flotation, the material is sent to **two-stage purification.**

```python
from IPython.display import display, Image

Image(url="https://pictures.s3.yandex.net/resources/viruchka_2_1589899167.jpg")
```

**1. Flotation**

Gold ore mixture is fed into the float banks to obtain **rougher Au concentrate** and **rougher tails** (product residues with a low concentration of valuable metals).
The stability of this process is affected by the volatile and non-optimal physicochemical state of the flotation pulp (a mixture of solid particles and liquid).

**2. Purification**

The rougher concentrate undergoes two stages of purification. After purification, we have the **final concentrate** and **new tails.**

### Data description

**Technological process**

- Rougher feed — raw material
- Rougher additions (or reagent additions) — flotation reagents: Xanthate, Sulphate, Depressant
    - Xanthate — promoter or flotation activator;
    - Sulphate — sodium sulphide for this particular process;
    - Depressant — sodium silicate.
- Rougher process — flotation
- Rougher tails — product residues
- Float banks — flotation unit
- Cleaner process — purification
- Rougher Au — rougher gold concentrate
- Final Au — final gold concentrate


**Parameters of stages**

- air amount — volume of air
- fluid levels
- feed size — feed particle size
- feed rate


### Feature naming

Here's how we name the features:
- [stage].[parameter_type].[parameter_name]
- **Example:** rougher.input.feed_ag

**Possible values for [stage]:**
- rougher — flotation
- primary_cleaner — primary purification
- secondary_cleaner — secondary purification
- final — final characteristics

**Possible values for [parameter_type]:**
- input — raw material parameters
- output — product parameters
- state — parameters characterizing the cur
- rent state of the stage
- calculation — calculation characteristics

### Recovery calculation

We need to simulate **the process of recovering gold from gold ore.**

We will use the following formula to simulate the recovery process:

```python
Image(url="https://pictures.s3.yandex.net/resources/Recovery_1576238822_1589899219.jpg")
```

- **C** — share of gold **in the concentrate right after flotation** (for finding the rougher concentrate recovery) **/** **after purification** (for finding the final concentrate recovery) 

- **F** — share of gold **in the feed before flotation** (for finding the rougher concentrate recovery) **/** **in the concentrate right after flotation** (for finding the final concentrate recovery) 

- **T** — share of gold **in the rougher tails right after flotation** (for finding the rougher concentrate recovery) **/** **after purification** (for finding the final concentrate recovery) 

To predict the coefficient, we need to find the share of gold in the concentrate and the tails. We need to note that **both final and rougher concentrates matter.**

### Evaluation metric

To solve the problem, we will need a new metric. It is called **sMAPE**, **symmetric Mean Absolute Percentage Error.**

It is **similar to MAE, but is expressed in relative values instead of absolute ones.** Why is it symmetrical? It equally takes into account the scale of both the target and the prediction.

Here’s how sMAPE is calculated:

```python
Image(url="https://pictures.s3.yandex.net/resources/smape_1576238825_1589899257.jpg")
```

### Check that recovery is calculated correctly. Using the training set, calculate recovery for the ***rougher.output.recovery*** feature. Find the MAE between your calculations and the feature values. Provide findings.

- We will calculate and check the rougher recovery.

**C: 'rougher.output.concentrate_au'**

**F: 'rougher.input.feed_au'**

**T: 'rougher.output.tail_au'**

```python
rougher_C = train_data['rougher.output.concentrate_au']
rougher_F = train_data['rougher.input.feed_au']
rougher_T = train_data['rougher.output.tail_au']
```

- We will store them in a list.

```python
rougher_recovery_parameters = [rougher_C, rougher_F, rougher_T]
```

- Let's check the missing values and zeros for recovery parameters.

```python
print("\n0 and NaN values for C, for F, and for T")
for r in rougher_recovery_parameters:
    print("\n|0   : {:<5}|".format((r == 0).sum()))
    print("|NaN : {:<5}|".format(r.isna().sum()))
```

- Now we will check the zeros in other parameters of recovery formula which can lead to error.

```python
(rougher_C - rougher_T == 0).sum()
```

```python
(rougher_F - rougher_T == 0).sum()
```

- We will calculate the rougher recovery and create a new Series.

```python
rougher_recovery_calculated = pd.Series(100 * ((rougher_C * (rougher_F - rougher_T)) / (rougher_F * (rougher_C - rougher_T))))
```

- Let's create a table from two rougher recoveries.

```python
rougher_recovery_table = pd.DataFrame(data={'rougher_output_recovery': train_data['rougher.output.recovery'], 
                                       'rougher_calculated_recovery': rougher_recovery_calculated})
```

```python
rougher_recovery_table.head()
```

- We will check the missing values, then drop to avoid calculation error.

```python
rougher_recovery_table.isna().sum()
```

```python
rougher_recovery_table.dropna(axis=0, inplace=True)
```

- Let's check again.

```python
rougher_recovery_table.isna().sum()
```

- Now we can calculate.

- We will use **sklearn.metrics.mean_absolute.error** to calculate MAE.

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(rougher_recovery_table['rougher_output_recovery'], rougher_recovery_table['rougher_calculated_recovery'])
```

- It's too low, and it means the recovery is calculated correctly.

- Let's take a look at the histograms of recoveries.

```python
plt.figure(figsize=(13,8))

plt.hist(rougher_recovery_table['rougher_output_recovery'], color='blue', alpha=.4, bins=100)
plt.hist(rougher_recovery_table['rougher_calculated_recovery'], color='red', alpha=.4, bins=100)

plt.grid(True, alpha=0.3, color='gray')

plt.title('\nRougher Recovery Distributions\n', size=21)
plt.xticks(range(0, 110, 10), size=15)
plt.yticks(size=15)
plt.xlabel('\nRecovery\n', size=18)
plt.ylabel('Distribution\n', size=18)
plt.legend(['feature recovery values', 'calculated recovery values'], fontsize=15);
```

#### Outliers Detection and Handling

- We will define a function to get a table to show outliers.

```python
def column_sigma_outlier(series, error):
    
    mean = series.mean()
    sigma = series.std()
        
    upp_lim = mean + error * sigma
    low_lim = mean - error * sigma
    
    lower_outliers = series[series < low_lim]
    upper_outliers = series[series > upp_lim]
        
    count_lowers = len(lower_outliers)
    count_uppers = len(upper_outliers)
        
    count_all = count_lowers + count_uppers
    rate_all = count_all / len(series) * 100
    
    outliers_data = [[low_lim, count_lowers, upp_lim, count_uppers, count_all, rate_all]]
    outliers_columns = ['lower_limit', 'lower_quantities', 'upper_limit', 'upper_quantities', 'all_quantities', 'rate_all']
    outliers_table = pd.DataFrame(data=outliers_data, columns=outliers_columns)
    return outliers_table
```

### Write a function to calculate the final sMAPE value

- Let's remember the sMAPE calculation.

```python
Image(url="https://pictures.s3.yandex.net/resources/smape_1576238825_1589899257.jpg")
```

- We will write this function.
- Let's look at the final sMAPE calculation.

```python
Image(url="https://pictures.s3.yandex.net/resources/_smape_1589899561.jpg")
```

```python
def s_mape(rougher_real_tgt, rougher_pred_tgt, final_real_tgt, final_pred_tgt):
    
    n = len(rougher_real_tgt)
    
    s_MAPE_rougher = ((abs(rougher_real_tgt - rougher_pred_tgt) / ((abs(rougher_real_tgt) + abs(rougher_pred_tgt)) / 2)) * 100).mean()
    s_MAPE_final = ((abs(final_real_tgt - final_pred_tgt) / ((abs(final_real_tgt) + abs(final_pred_tgt)) / 2)) * 100).mean()

    s_MAPE = .25 * s_MAPE_rougher + .75 * s_MAPE_final
    return s_MAPE
```

#### Linear Regression

- First we will look at the Linear Regression model.
- We will define a function to get sMAPE value.

```python
def linear_regression_smape(features1, target1, features2, target2):

    model = LinearRegression()
    model.fit(features1, target1)
    preds1 = model.predict(features1)
    model.fit(features2, target2)
    preds2 = model.predict(features2)
        
    s_MAPE = s_mape(target1, preds1, target2, preds2)
    print("|sMAPE: {:.2f}|".format(s_MAPE))
        
    return pd.to_numeric('{:.2f}'.format(s_MAPE))
```

#### Decision Tree Regressor

- We will try to find the best value for 'max_depth' parameter. 

```python
def tree_max_depth(features1, target1, features2, target2):
    
    depth_sMAPE_dict = dict()
    
    for depth in range(1, 21):

        model = DecisionTreeRegressor(random_state=99, max_depth=depth)
        model.fit(features1, target1)
        preds1 = model.predict(features1)
        model.fit(features2, target2)
        preds2 = model.predict(features2)
        
        s_MAPE = s_mape(target1, preds1, target2, preds2)
        depth_sMAPE_dict[depth] = s_MAPE
        print("|Max depth: {: <3} |sMAPE: {:.2f}|".format(depth, s_MAPE))
        
        
    max_depth = min(depth_sMAPE_dict, key=depth_sMAPE_dict.get)
    min_s_mape = depth_sMAPE_dict[max_depth]
    return max_depth, pd.to_numeric('{:.2f}'.format(min_s_mape))
```

```python
def forest_max_depth(features1, target1, features2, target2):
    
    depth_sMAPE_dict = dict()
    
    for depth in range(2, 21, 2):

        model = RandomForestRegressor(random_state=99, max_depth=depth, n_estimators=20)
        model.fit(features1, target1)
        preds1 = model.predict(features1)
        model.fit(features2, target2)
        preds2 = model.predict(features2)
        
        s_MAPE = s_mape(target1, preds1, target2, preds2)
        depth_sMAPE_dict[depth] = s_MAPE
        print("|Max depth: {: <3}|sMAPE: {:.2f}|".format(depth, s_MAPE))
        
    max_depth = min(depth_sMAPE_dict, key=depth_sMAPE_dict.get)
    min_s_mape = depth_sMAPE_dict[max_depth]
    return max_depth, pd.to_numeric('{:.2f}'.format(min_s_mape))
```

```python
def forest_n_estim(features1, target1, features2, target2, depth=20):
    
    n_sMAPE_dict = dict()
    
    for n_estim in range(10, 41, 5):

        model = RandomForestRegressor(random_state=99, max_depth=depth, n_estimators=n_estim)
        model.fit(features1, target1)
        preds1 = model.predict(features1)
        model.fit(features2, target2)
        preds2 = model.predict(features2)
        
        s_MAPE = s_mape(target1, preds1, target2, preds2)
        n_sMAPE_dict[n_estim] = s_MAPE
        print("|n_estim: {: <3} |sMAPE: {:.2f}|".format(n_estim, s_MAPE))
        
    n_estim = min(n_sMAPE_dict, key=n_sMAPE_dict.get)
    min_s_mape = n_sMAPE_dict[n_estim]
    return n_estim, pd.to_numeric('{:.2f}'.format(min_s_mape))
```

#### Cross Validation

- We will use 5 as 'cv fold' to divide datasets into train and validation. (80-20 %)
- We will use divide sets with for loop in the 'cross_validation' function.

```python
def cross_validation(model, features1, target1, features2, target2):
    
    smape_list = []
    size = len(features1)
    sample_size = len(features1) // 5
    total = 0

    for i in range(0, size, sample_size):
        
        if (i+sample_size > size):
            valid_indexes = range(i, size)
            train_indexes = [i for i in range(i)]
        else:
            valid_indexes = range(i, i+sample_size)
            train_indexes = [j for k in (range(i), range(i+sample_size, size)) for j in k]
        
        features1_train = features1.iloc[train_indexes]
        target1_train = target1.iloc[train_indexes]
        features1_valid = features1.iloc[valid_indexes]
        target1_valid = target1.iloc[valid_indexes]

        features2_train = features2.iloc[train_indexes]
        target2_train = target2.iloc[train_indexes]
        features2_valid = features2.iloc[valid_indexes]
        target2_valid = target2.iloc[valid_indexes]
        
        model.fit(features1_train, target1_train)
        preds1 = model.predict(features1_valid)
        
        model.fit(features2_train, target2_train)
        preds2 = model.predict(features2_valid)
        
        s_MAPE = s_mape(target1_valid, preds1, target2_valid, preds2)
        smape_list.append(s_MAPE)
        
        total += s_MAPE
        
    final_sMAPE = total / 5

    return pd.to_numeric('{:.3f}'.format(final_sMAPE))
```
